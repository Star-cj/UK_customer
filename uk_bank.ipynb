```python
# Number of people within an age bracket 
children=df[df.Age<18].count()
age_18_25=df[(df['Age']>=18) & (df['Age']<=25)].count()
age_26_35=df[(df['Age']>25) & (df['Age']<=35)].count()
age_36_45=df[(df['Age']>35) & (df['Age']<=45)].count()
age_46_55=df[(df['Age']>45) & (df['Age']<=55)].count()
age_56_65=df[(df['Age']>55) & (df['Age']<=65)].count()

children
# age_36_45['Age']
```

```python
x=np.array([children['Age'],age_18_25['Age'],age_26_35['Age'],age_36_45['Age'],age_46_55['Age'],age_56_65['Age']])
lable=['<18','>=18 & <=25','>25 & <=35','>35 & <=45','>45 & <=55','>55 & <=65']
plt.title('Total Age bracket of customer',size=20)

plt.pie(x,labels=lable,autopct='%1.1f%%')
# plt.legend(loc='upper right', title='key')     #the legend key is too long to block many information
plt.show()
```

To check age bracket of customers based on their gender (male or female)  

```python
# will use for both male and female

children=df[df.Age<18]
age_18_25=df[(df['Age']>=18) & (df['Age']<=25)]
age_26_35=df[(df['Age']>25) & (df['Age']<=35)]
age_36_45=df[(df['Age']>35) & (df['Age']<=45)]
age_46_55=df[(df['Age']>45) & (df['Age']<=55)]
age_56_65=df[(df['Age']>55) & (df['Age']<=65)]

# d=age_18_25[(age_18_25.Gender=='Female')]
# d=female18_25[(female18_25.Gender=='Female')].count()
# d.Gender.count()
```

```python
maleChild =children [(children.Gender=='Male')].count()
male18_25=age_18_25[(age_18_25.Gender=='Male')].count()
male26_35=age_26_35[(age_26_35.Gender=='Male')].count()
male36_45=age_36_45[(age_36_45.Gender=='Male')].count()
male46_55=age_46_55[(age_46_55.Gender=='Male')].count()
male56_65=age_56_65[(age_56_65.Gender=='Male')].count()
maleChild.Gender
```

```python
# compare the age bracket of each gender (male)

x=np.array([maleChild.Gender,male18_25.Gender,male26_35.Gender,male36_45.Gender,male46_55.Gender,male56_65.Gender])
lable=['male <18','male >=18 & <=25','male >25 & <=35','male >35 & <=45','male >45 & <=55','male >55 & <=65']
plt.title('Age bracket of Male',size=20)

plt.pie(x,labels=lable,autopct='%1.1f%%')
plt.show()
```

```python
femaleChild =children [(children.Gender=='Female')].count()
female18_25=age_18_25[(age_18_25.Gender=='Female')].count()
female26_35=age_26_35[(age_26_35.Gender=='Female')].count()
female36_45=age_36_45[(age_36_45.Gender=='Female')].count()
female46_55=age_46_55[(age_46_55.Gender=='Female')].count()
female56_65=age_56_65[(age_56_65.Gender=='Female')].count()
femaleChild.Gender
```

```python
# compare the age bracket of each gender (female)

x=np.array([femaleChild.Gender,female18_25.Gender,female26_35.Gender,female36_45.Gender,female46_55.Gender,female56_65.Gender])
lable=['female <18','female >=18 & <=25','female >25 & <=35','female >35 & <=45','female >45 & <=55','female >55 & <=65']
plt.title('Age bracket of Female',size=20)

plt.pie(x,labels=lable,autopct='%1.1f%%')
plt.show()
```

```python
# compare both male and female 
gender={
        'Male':[maleChild.Gender,male18_25.Gender,male26_35.Gender,male36_45.Gender,male46_55.Gender,male56_65.Gender],
        'Female':[femaleChild.Gender,female18_25.Gender,female26_35.Gender,female36_45.Gender,female46_55.Gender,female56_65.Gender],
        'Age Bracket':['<18','>=18 & <=25','>25 & <=35','>35 & <=45','>45 & <=55','>55 & <=65']
}
genderData=pd.DataFrame(gender)
genderData
```

```python
maledata=genderData.Male.sum()
femaledata=genderData.Female.sum()
x=[maledata,femaledata]
gen_label=["Male","Female"]
print('There are '+str(maledata)+' Male;',str(femaledata)+' Female')
plt.pie(x=x,labels=gen_label,autopct='%1.1f%%')
plt.title("Different gender that entered Uk bank")
```

```python
# comparing the genders using line graph

genderData.plot(x='Age Bracket',title="Compare costumer based on gender", colormap='plasma')  #colormap='hsv' ,'cool', 'viridis'has many other colors
plt.xticks(rotation=90)
```
